## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 1. Total pages and baseline declining rate
total_rows = df.shape[0]
declining_rate = df["trend_direction"].str.lower().eq("down").mean()
print(f"Total pages in dataset: {total_rows}")
print(f"Base declining rate: {declining_rate:.3f} ({round(declining_rate*100, 1)}% of pages are trending down)")

# 2. Content Age vs Decline (supporting number)
age_median = df.groupby("trend_direction")["content_age_days"].median()
print("\nMedian content age by trend direction:")
print(age_median.to_string())
```
