## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
Classification.

We want to predict a binary outcome: whether a job applicant will get shortlisted or not (Yes/No).
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
The Target is is_shortlisted (values 0 or 1).

This label comes from observed outcomes—meaning actual historical hiring decisions recorded on the platform, not a manual if-else rule.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
F1-Score.

Since most applicants do not get shortlisted, the dataset is imbalanced. A simple accuracy metric can be misleading. An F1-Score of 0.80 or higher is our target for a good model.
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

# Standard sample data representing our data slice
data = {
    'candidate_id': ['FR-901', 'FR-902', 'FR-903'],
    'coding_score': [85, 92, 60],
    'experience_years': [3, 5, 1],
    'communication_rating': [4.5, 4.8, 3.2],
    'is_shortlisted': [1, 1, 0]
}

df = pd.DataFrame(data)
print("Unit of analysis: One row represents a single Job Applicant's profile details.")
df.head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
A simple manual rule like if coding_score > 80 and experience_years > 3: shortlist is too rigid.

It will reject outstanding candidates who have excellent coding skills but slightly less experience, or candidates with brilliant communication skills. Machine learning beats fixed rules because it learns the complex, non-linear relationships between all these features automatically.
```
