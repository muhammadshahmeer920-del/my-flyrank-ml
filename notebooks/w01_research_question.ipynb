## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

corr = df["search_volume"].corr(df["impressions_90d"])
print(f"1) Correlation search_volume vs impressions_90d: {corr:.3f} (near zero)")

visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print("\n2) Mean CTR by position tier (impressions >= 100):")
print(ctr_by_pos.round(3))

wc = df.groupby("trend_direction")["word_count"].median()
print("\n3) Median word count by trend_direction:")
print(wc)

print("\nEffect size: page_1 CTR (~0.35) vs deep CTR (~0.06) is roughly a 6x gap —")
print("a large practical effect, not just a statistically detectable one.")
```
